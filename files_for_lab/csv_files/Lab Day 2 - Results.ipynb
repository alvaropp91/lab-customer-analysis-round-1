```python
# Standarize header names file1 - lower case

cols1 = []
for colname in file1.columns:
    cols1.append(colname.lower())
file1.columns = cols1
```

```python
display (file1)
```

```python
# Standarize header names file2 - lower case

cols2 = []
for colname in file2.columns:
    cols2.append(colname.lower())
file2.columns = cols2
```

```python
display (file2)
```

```python
# Standarize header names file3 - lower case

cols3 = []
for colname in file3.columns:
    cols3.append(colname.lower())
file3.columns = cols3
```

```python
# Rearrange columns in file2

file2=file2[["st","gender","education","customer lifetime value", "income", "monthly premium auto", "number of open complaints", "policy type", "vehicle class", "total claim amount"]]
```

```python
display(file2)
```

```python
#Rename columns in file3 and rearrenge columns in file3

file3 = file3.rename(columns={'state':'st'})
file3=file3[["st","gender","education","customer lifetime value", "income", "monthly premium auto", "number of open complaints", "policy type", "vehicle class", "total claim amount"]]
```

```python
display(file3)
```

```python
# Delete % from file 1
file1["customer lifetime value"] = file1["customer lifetime value"].str.replace("%","")
file1
```

```python
# Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)
# In file 1
file1['customer lifetime value'] =  pd.to_numeric(file1['customer lifetime value'], errors='coerce')/100
file1
```

```python
# Delete % from file 2
file2["customer lifetime value"] = file2["customer lifetime value"].str.replace("%","")
file2
```

```python
# In file 2
file2['customer lifetime value'] =  pd.to_numeric(file2['customer lifetime value'], errors='coerce')/100
```

```python
file2
```

```python
#Concatenate all files
data = pd.concat([file1,file2,file3], axis=0)
```

```python
#Which columns are numerical?

data._get_numeric_data()
```

```python
#Which columns are categorical?
data.select_dtypes('object')
```

```python
#Delete the column education and the number of open complaints from the dataframe.

data.drop (["education","number of open complaints"], axis=1, inplace=True)
```

```python
#Check for duplicate rows in the data and remove if any.
data = data.drop_duplicates()
```

```python
data
```

```python
#Filter out the data for customers who have an income of 0 or less. (Only show customers who HAVE an income.)
data_filtered = data[(data['income']>0)]
data = data.drop(data_filtered, axis=1)
```
